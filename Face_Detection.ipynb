```python
#using haar cascade for face detection
```

```python
cascade=cv2.CascadeClassifier(r'C:\Users\dhanu\Desktop\EDU\OpenCV\Computer-Vision-with-Python\DATA\haarcascades\haarcascade_frontalface_default.xml')
```

```python
def detectface(img):
    copy=img.copy()
    final=cascade.detectMultiScale(copy, scaleFactor=1.2, minNeighbors=5)           #gets the value of coordinates
    
    for(a,b,c,d) in final:       
        cv2.rectangle(copy, (a,b), (a+c,b+d), (255,255,255), 4)
        
    return copy
```

```python
plt.imshow(detectface(img), cmap='gray')
```

```python
plt.imshow(detectface(group), cmap='gray')
```
